```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import PLSCanonical
from sklearn.cross_decomposition import CCA
from sklearn.cross_decomposition import PLSSVD
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
data = pd.read_json('./data.json')
```

```python
X = data.x.values.reshape(-1,1)
y = data.y.values
```

```python
pls2 = PLSRegression(n_components=1)
pls2.fit(X, y)
y_pred = pls2.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title("PLSRegression", fontsize=16)
ax.set_ylabel("y", fontsize=14)
ax.set_xlabel("x", fontsize=14)
ax.scatter(X, y_pred, color='red', label='predict')
ax.scatter(X, y,  color='black', label='origin')
ax.legend()
plt.show()
```

```python
pls2 = PLSCanonical(n_components=1)
pls2.fit(X, y)
y_pred = pls2.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title("PLSCanonical", fontsize=16)
ax.set_ylabel("y", fontsize=14)
ax.set_xlabel("x", fontsize=14)
ax.scatter(X, y_pred, color='red', label='predict')
ax.scatter(X, y,  color='black', label='origin')
ax.legend()
plt.show()
```

```python
pls2 = CCA(n_components=1)
pls2.fit(X, y)
y_pred = pls2.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title("CCA", fontsize=16)
ax.set_ylabel("y", fontsize=14)
ax.set_xlabel("x", fontsize=14)
ax.scatter(X, y_pred, color='red', label='predict')
ax.scatter(X, y,  color='black', label='origin')
ax.legend()
plt.show()
```
