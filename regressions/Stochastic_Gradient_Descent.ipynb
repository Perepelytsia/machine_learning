```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

plt.style.use('ggplot')
data = pd.read_json('./data.json')
```

```python
X = data.x.values.reshape(-1,1)
y = data.y.values
```

```python
reg = SGDRegressor(max_iter=1000, tol=1e-3)
reg.fit(X, y)
y_pred = reg.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title("Stochastic Gradient Descent without scaling", fontsize=16)
ax.set_ylabel("y_origin, y_ped", fontsize=14)
ax.set_xlabel("x", fontsize=14)
ax.scatter(X, y_pred, color='red', label='predict')
ax.scatter(X, y,  color='black', label='origin')
ax.legend()
plt.show()
```

```python
est = make_pipeline(StandardScaler(), SGDRegressor())
est.fit(X, y)
y_pred = est.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title("Stochastic Gradient Descent with scaling", fontsize=16)
ax.set_ylabel("y_origin, y_ped", fontsize=14)
ax.set_xlabel("x", fontsize=14)
ax.scatter(X, y_pred, color='red', label='predict')
ax.scatter(X, y,  color='black', label='origin')
ax.legend()
plt.show()
```
