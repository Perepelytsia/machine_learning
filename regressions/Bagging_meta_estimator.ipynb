```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.ensemble import BaggingRegressor

plt.style.use('ggplot')
data = pd.read_json('./data.json')
```

```python
X = data.x.values.reshape(-1,1)
y = data.y.values
```

```python
regressor = svm.SVR()
regressor.fit(X, y)
y_pred = regressor.predict(X)
```

```python
regr2 = BaggingRegressor(base_estimator=svm.SVR(), n_estimators=50, random_state=1)
regr2.fit(X, y)
y_pred2 = regr2.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_title("Support Vector Regression", fontsize=2)
ax.set_ylabel("y", fontsize=14)
ax.set_xlabel("x", fontsize=14)
ax.scatter(X, y_pred, color='red', label='support vector regression')
ax.scatter(X, y,  color='black', label='origin')
ax.scatter(X, y_pred2,  color='blue', label='bagging')
ax.legend()
plt.show()
```
