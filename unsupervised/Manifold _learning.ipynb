```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import manifold
```

```python
x = np.random.randint(10, 16, 50)
y = np.random.randint(11, 15, 50)
z = np.random.randint(5, 20, 50)
X = x.reshape(50,1)
Y = y.reshape(50,1)
Z = z.reshape(50,1)
data = np.concatenate((X, Y, Z), axis=1)
```

```python
fig = plt.figure(figsize=(13, 10))
ax = plt.axes(projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c='red', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim3d(5, 20)
ax.set_ylim3d(5, 20)
ax.set_zlim3d(5, 20)
plt.show()
```

```python
def result(model):
    embedding = model(n_components=2)
    data_transformed = embedding.fit_transform(data)
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.scatter(data_transformed[:, 0], data_transformed[:, 1], color="red")
    plt.show()
```

```python
result(manifold.Isomap)
```

```python
result(manifold.LocallyLinearEmbedding)
```

```python
result(manifold.SpectralEmbedding)
```

```python
result(manifold.MDS)
```

```python
result(manifold.TSNE)
```
